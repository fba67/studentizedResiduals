```r
set.seed(1)
cell <- "K562" ## can either be HepG2 or K562
```

```r
if(cell == "HepG2"){
  genes.TPM <- read.table(paste("~/mount2/singleCell/", cell, "/genesMat.TPM.160923_ENSGRsRemoved.txt", sep= ""), header= T, stringsAsFactors= F)
  hm <- read.table("~/mount2/DEEP_ChIP_HM/01/HepG2/01_HepG2_allGenes__2000bp.x")
}else{
  genes.TPM <- read.table(paste("~/mount2/singleCell/", cell, "/K562_genesMat.TPM.170302_ENSGRsRemoved.txt", sep= ""), header= T, stringsAsFactors= F)
  hm <- read.table("~/mount2/ENC_ChIP_HM/plus/K562/K562_allGenes.x")
}

genes.TPM <- data.frame(gene_id= genes.TPM$gene_id, avg_expr= rowMeans(genes.TPM[, -1]), stringsAsFactors= F)
```

```r
library(biomaRt)

ens.ids <- sapply(seq(nrow(genes.TPM)), function(i) unlist(strsplit(genes.TPM$gene_id[i], "\\."))[1])
genes.TPM$gene_id <- ens.ids

ensembl = useMart(biomart="ENSEMBL_MART_ENSEMBL", host="grch37.ensembl.org", path="/biomart/martservice" ,dataset="hsapiens_gene_ensembl")
biomartres <- getBM(attributes=c("start_position", "end_position", "ensembl_gene_id", "chromosome_name", "strand"), values= ens.ids, mart= ensembl)

valid.chrs <- c("X", "Y", as.character(seq(22)))
valid.chr.hits <- which(biomartres$chromosome_name %in% valid.chrs)
biomartres.valid <- biomartres[valid.chr.hits, ]

hits <- which(biomartres.valid$ensembl_gene_id %in% ens.ids)
hit.genes <- biomartres.valid[hits, ]

### filter the short genes of length smaller than 2000bp
long.genes.idx <- which(abs(hit.genes$start_position - hit.genes$end_position) >= 2000)
long.genes <- hit.genes[long.genes.idx, ]

### correct for the minus strand TSS
minus.strand.idx <- which(long.genes$strand == -1)
temp <- long.genes$start_position[minus.strand.idx]
long.genes$start_position[minus.strand.idx] <- long.genes$end_position[minus.strand.idx]
long.genes$end_position[minus.strand.idx] <- temp

sorted.long.genes <- long.genes[order(long.genes$chromosome_name, long.genes$start_position, decreasing= F), ]
keep.idx<- NULL;
for(i in seq(nrow(sorted.long.genes) - 1)){
  diff <- sorted.long.genes$start_position[i] + 2000 - (sorted.long.genes$start_position[i + 1] - 2000);
  if(diff > 0 && sorted.long.genes$chromosome_name[i] == sorted.long.genes$chromosome_name[i + 1]){
    keep.idx <- c(keep.idx, i)
  }else if(diff < 0){
    keep.idx <- c(keep.idx, i)
  }
}

final.genes <- sorted.long.genes[keep.idx, ]
final.genes.somatic <- final.genes[which(final.genes$chromosome_name %in% valid.chrs[-c(1,2)]), ]
final.genes.germ <- final.genes[which(final.genes$chromosome_name %in% valid.chrs[c(1,2)]), ]

final.genes.somatic$chromosome_name <- as.numeric(final.genes.somatic$chromosome_name)
final.genes.somatic <- final.genes.somatic[order(final.genes.somatic$chromosome_name, final.genes.somatic$start_position, decreasing= F), ]
final.genes.germ <- final.genes.germ[order(final.genes.germ$chromosome_name, final.genes.germ$start_position, decreasing= F), ]

final.genes.sorted <- rbind(final.genes.somatic, final.genes.germ)
colnames(final.genes.sorted)[3] <- "gene_id"
                  
hm <- cbind(gene_id= final.genes.sorted$gene_id, hm)
data  <- merge(genes.TPM, hm, by= "gene_id") ## contains both hm and expr data

```

```r
print(dim(data))
```

```r
get.rss <- function(pred, y){
  n <- length(y)
  rmse <- sqrt(1 / n * sum((pred - y) ^ 2))
  return(rmse)
}
```

```r
data.x <- log2(1 + data[, -c(1,2)])
data.y <- log2(1 + data[, 2])
```

```r
## OLS
library(MASS)

lm.res <- lm(data.y ~ as.matrix(data.x), y= T, x= T)
studres.res <- studres(lm.res);

stable.idx <- which(studres.res > 1)
unstable.idx <- which(studres.res < -1)

cols <- rep("black", length(studres.res));
cols[stable.idx] <- "red";
cols[unstable.idx] <- "blue";
```

```r
plot(lm.res$y, studres.res, pch= 20, col= cols, main= cell, xlab= "measured mRNA", ylab= "Studentized residuals")
abline(h= c(1, -1))
```

```r
rmse <- round(get.rss(lm.res$fitted.values, lm.res$y), 2)
cor_p <- round(cor(lm.res$fitted.values, lm.res$y), 2)
cor_s <- round(cor(lm.res$fitted.values, lm.res$y, method= "spearman"), 2)
```

```r
plot(lm.res$fitted.values, lm.res$y, pch= 20, main= paste("RMSE=", rmse, "cor_s=", cor_s, "cor_p=", cor_p))
abline(lm(lm.res$fitted.values ~ lm.res$y), col= "red")
```
